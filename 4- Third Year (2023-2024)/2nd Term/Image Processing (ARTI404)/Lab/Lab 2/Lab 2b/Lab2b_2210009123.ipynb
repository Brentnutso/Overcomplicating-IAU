```python
# Task 1: Importing and Cropping
import cv2
from google.colab.patches import cv2_imshow
import sys

try:
  image = cv2.imread("mario.jpg", cv2.IMREAD_COLOR)
except:
  print("Unable to open the image")
  sys.exit(1)

# original image
# cv2_imshow(image)

# cropping
crop_x, crop_y, crop_width, crop_height = 0, 0, 750, 550
cv2.rectangle(image, (crop_x, crop_y), (crop_x + crop_width, crop_y + crop_height), (0, 255, 0), 2)
crop = image[crop_y:crop_y+crop_height, crop_x:crop_x+crop_width]

cv2_imshow(crop)
```

```python
# Task 2: Rotating
rotated = cv2.rotate(crop,cv2.ROTATE_180)
cv2_imshow(rotated)
```

```python
#Task 3: Make gray-scale
grayscale = cv2.cvtColor(crop, cv2.COLOR_BGR2GRAY)
cv2_imshow(grayscale)
```
