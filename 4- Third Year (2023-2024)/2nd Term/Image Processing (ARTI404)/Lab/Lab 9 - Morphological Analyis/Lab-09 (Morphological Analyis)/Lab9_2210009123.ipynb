```python
# Imports
import matplotlib.pyplot as plt
from skimage.util import img_as_ubyte
from skimage.color import rgb2gray
from skimage.morphology import dilation, erosion, disk
from skimage.transform import resize
import os


img1 = rgb2gray(img_as_ubyte(plt.imread('images/Picture1.png')[:, :, :3]))
img2 = rgb2gray(img_as_ubyte(plt.imread('images/Picture2.png')[:, :, :3]))


new_size = (img1.shape[0]*3, img1.shape[1]*3)

img1_resized = img_as_ubyte(resize(img1, new_size))
img2_resized = img_as_ubyte(resize(img2, new_size))
```

```python
# Task 1: Dilation
img1_dilated = dilation(img1_resized, disk(7))
# Show the result along with the original image
plt.figure()
plt.subplot(1, 2, 1)
plt.imshow(img1_resized, cmap='gray')
plt.title('Original Image')
plt.subplot(1, 2, 2)
plt.imshow(img1_dilated, cmap='gray')
plt.title('Dilated Image')
plt.show()

# Save the result
if not os.path.exists('results'):
    os.makedirs('results')
plt.imsave('results/img1_original.png', img1_resized, cmap='gray')
plt.imsave('results/img1_dilated.png', img1_dilated, cmap='gray')
```

```python
# Task 2: Erosion
img2_erosion = erosion(img2_resized, disk(10))

# Show the result along with the original image
plt.figure()
plt.subplot(1, 2, 1)
plt.imshow(img2_resized, cmap='gray')
plt.title('Original Image')
plt.subplot(1, 2, 2)
plt.imshow(img2_erosion, cmap='gray')
plt.title('Erosion Image')
plt.show()

# Save the result
if not os.path.exists('results'):
    os.makedirs('results')
plt.imsave('results/img2_original.png', img2_resized, cmap='gray')
plt.imsave('results/img2_erosion.png', img2_erosion, cmap='gray')
```
